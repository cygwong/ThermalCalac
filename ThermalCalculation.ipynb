```python
#numerically integrate- Riemann Sum- the MB distribution

import numpy as np


#constants

k=1.38e-23


M=float(input ('mass of atom in amu'))

m=M*1.67e-27

T=float(input ('Kelvin temperature'))

v1=float(input ('lower speed'))

v2=float(input ('upper speed'))

N=float(input ('number of rectangles'))




def MB(v):

    PDF=4./np.sqrt(np.pi)*(m/2./k/T)**(1.5)*v**2.*np.exp(-m*v**2/2./k/T)

    return PDF

dv=(v2-v1)/N #this is the base or stepsize to make N rectangles between v1 and v2

v=v1 #start at v1

sum=0

while v<=v2:

    sum=sum+MB(v)*dv #sum is the area under the MB curve, height * base; call the MB function

    v=v+dv

print ('Riemann sum of the Maxwell Boltzmann Probability Distribution function(PDF) is',sum)

print (np.sqrt(2*k*T/m))


print (sum*100.,'% of molecules have a speed in this range')
```
