```python
import numpy as np

gasR=8.314

gamma=1.67 

T1 = 288.
T2 = 260.


#first find P,V,T using ideal gas law

TC=T1 #kpa L units

VC=0.001

PC=1.0*101325 #pascal

n= PC*VC/(gasR*TC)


TD=T2

VD=(TC*VC**(gamma-1)/TD)**(1/(gamma-1))

PD=PC*(VC/VD)**(gamma)


TA=T2

VA= 0.00095

PA=n*gasR*TA/VA


TB= T1

VB=(TA*VA**(gamma-1)/TB)**(1/(gamma-1)) #adiabatic

PB=PA*(VA/VB)**(gamma)


print('Temp ABCD',TA,TB,TC, TD)

print('Pressure ABCD',PA, PB, PC, PD)

print('Volume ABCD',VA, VB, VC, VD)

#adiabatic
QAB=0

WAB=-PA*VA**gamma*(VB**(1-gamma)-VA**(1-gamma))/(1-gamma) #integrate adiabatic

DUAB=WAB+QAB

#ISOTHERMAL
WBC=n*gasR*TC*np.log(VC/VB)

QBC=-WBC

DUBC=QBC+WBC

#adiabatic
QCD=0

WCD=-PC*VC**gamma*(VD**(1-gamma)-VC**(1-gamma))/(1-gamma) #integrate adiabatic

DUCD=WCD+QCD

#ISOTHERMAL
WDA=n*gasR*TA*np.log(VA/VD)

QDA=-WDA

DUDA=QDA+WDA


overallW=WAB+WBC+WCD+WDA

overallQ=QAB+QBC+QCD+QDA

overallDU=DUAB+DUBC+DUCD+DUDA


print ('AB work, heat, DU',WAB,QAB,DUAB)

print ('BC work, heat, DU',WBC,QBC,DUBC)

print ('CD work, heat, DU',WCD,QCD,DUCD)

print ('DA work, heat, DU',WDA,QDA,DUDA)

print ('overall work, heat, DU',overallW,overallQ,overallDU)


eff=(overallW/QDA*100)

print('Efficentcy: ',eff)
```
